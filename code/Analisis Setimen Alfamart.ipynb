## Preprocessing

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory,ArrayDictionary
import re
from googletrans import Translator

# Lowercasing
dataframe['komentar'] = dataframe['komentar'].str.lower()


# Membuat objek penerjemah
translator = Translator()

# Fungsi untuk menerjemahkan teks
def translate_text(text):
    # Menerjemahkan teks ke bahasa Indonesia
    translated = translator.translate(text, dest='id')
    return translated.text

# Menerapkan fungsi ke kolom 'komentar'
dataframe['komentar'] = dataframe['komentar'].apply(translate_text)


# Normalisasi emoji dan kata-kata singkat
def normalize_text(text):
    # Normalisasi emoji
    text = re.sub(r':\)', ' senang ', text)
    text = re.sub(r':\(', ' sedih ', text)
    # Normalisasi kata-kata singkat
    text = re.sub(r'\bga\b', 'tidak', text)
    text = re.sub(r'\btdk\b', 'tidak', text)
    text = re.sub(r'\bgak\b', 'tidak', text)
    text = re.sub(r'\bkagak\b', 'tidak', text)
    text = re.sub(r'\bnggak\b', 'tidak', text)
    text = re.sub(r'\btp\b', 'tetapi', text)
    text = re.sub(r'\bgk\b', 'tetapi', text)
    text = re.sub(r'\blama\b', 'lambat', text)
    text = re.sub(r'\blamaa\b', 'lambat', text)
    text = re.sub(r'\byg\b', 'yang', text)
    text = re.sub(r'\btapi\b', 'tetapi', text)
    text = re.sub(r'\bsy\b', 'saya', text)
    text = re.sub(r'\blemot\b', 'lambat', text)
    text = re.sub(r'\bgjelas\b', 'tidak jelas', text)
    text = re.sub(r'\bgabisa\b', 'tidak bisa', text)
    text = re.sub(r'\bbgt\b', 'sekali', text)
    text = re.sub(r'\brisih\b', 'risi', text)


    return text

dataframe['komentar'] = dataframe['komentar'].apply(normalize_text)
```

```python
# Stemming
factory = StemmerFactory()
stemmer = factory.create_stemmer()
dataframe['komentar'] = dataframe['komentar'].apply(lambda x: stemmer.stem(x))

# Stopword removal
# Create stopword remover
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

# Get the list of stopwords
stopword_list = factory.get_stop_words()

# If 'tidak' is in the list, remove it
if 'tidak' in stopword_list:
    stopword_list.remove('tidak')

# Define a function to remove stopwords from a sentence
def remove_stopwords(sentence):
    words = sentence.split()
    words = [word for word in words if word not in stopword_list]
    return ' '.join(words)

# Apply the function to the 'komentar' column
dataframe['komentar'] = dataframe['komentar'].apply(remove_stopwords)

# Noise removal
dataframe['komentar'] = dataframe['komentar'].apply(lambda x: re.sub(r'\W', ' ', x))
```

```python
import pandas as pd
from textblob import TextBlob

# Misalkan kasir adalah DataFrame Anda
# kasir['komentar'] adalah kolom yang ingin Anda analisis sentimennya

# Kamus sentimen Anda
kamus_positif = {
    'jujur': 1,
    'ramah': 1,
    'sopan': 1,
    'baik': 1,
    'bagus': 1,
    'lengkap':1,
    'jelas' : 1,
    'bersih' : 1,
    'cepat':1,
    'murah':1,
}
kamus_negatif = {
    'lambat': -1,
    'buruk': -1,
    'bodoh': -1,
    'songong': -1,
    'kecewa' : -1,
    'jutek' : -1,
    'belagu' : -1,
    'risi' : -1,
    'busuk' : -1,
    'mahal':-1,
}


def hitung_sentimen(teks):
    kata_kata = teks.split()
    skor = 0
    for i in range(len(kata_kata)):
        if kata_kata[i] in kamus_positif:
            if (i > 0 and kata_kata[i-1] == 'tidak') or (i < len(kata_kata) - 1 and kata_kata[i+1] == 'tidak'):
                skor += -1
            else:
                skor += kamus_positif[kata_kata[i]]
        elif kata_kata[i] in kamus_negatif:
            skor += kamus_negatif[kata_kata[i]]
            if i < len(kata_kata) - 1 and kata_kata[i+1] == 'sekali':
                skor -= 1
            elif i > 0 and kata_kata[i-1] in ['sangat', 'paling'] : 
                skor -= 1
    return skor


def klasifikasi_sentimen(skor):
    if skor < 0:
        return 'negatif'
    elif skor > 0:
        return 'positif'
    else:
        return 'netral'

# Tambahkan kolom 'sentiment' dan 'klasifikasi' ke DataFrame
kasir['sentiment'] = kasir['komentar'].apply(hitung_sentimen)
kasir['klasifikasi'] = kasir['sentiment'].apply(klasifikasi_sentimen)
```

```python
jumlah_netral = kasir[kasir['klasifikasi'] == 'netral']['klasifikasi'].count()
jumlah_negatif = kasir[kasir['klasifikasi'] == 'negatif']['klasifikasi'].count()
jumlah_positif = kasir[kasir['klasifikasi'] == 'positif']['klasifikasi'].count()

# Total jumlah kata
total_kata = jumlah_netral + jumlah_negatif + jumlah_positif

# Menghitung persentase masing-masing jenis kata
persentase_netral = (jumlah_netral / total_kata) * 100
persentase_negatif = (jumlah_negatif / total_kata) * 100
persentase_positif = (jumlah_positif / total_kata) * 100

# Menampilkan hasil
print("Jumlah kata netral:", jumlah_netral)
print("Jumlah kata negatif:", jumlah_negatif)
print("Jumlah kata positif:", jumlah_positif)
print("\nPersentase kata netral: {:.2f}%".format(persentase_netral))
print("Persentase kata negatif: {:.2f}%".format(persentase_negatif))
print("Persentase kata positif: {:.2f}%".format(persentase_positif))
```
